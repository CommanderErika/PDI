## **3. Translate an image**

We know that a image it is basically an two dimensional matrix, so we can just change each pixel's index to make a translation.
To help us to do this task we'll be using numpy as a tool.

```python
# Getting image in gray scale #
img = cv.imread('images/sukuna.jpg', 0)

# Getting img shape #
shape = np.shape(img)
print("Image Height: {}\nImage Width: {}".format(shape[0], shape[1]))
```

```python
# Creating a function to translate image #
def TranslateImage(x, y, img):
    
    shape = np.shape(img)
    
    for i in range(shape[0]-x-1, 0, -1):
        for j in range(shape[1]-y-1, 0, -1):
            img[i+x][j+y] = img[i][j]

    for i in range(0, shape[0]):
        for j in range(0, shape[1]):
            if(i <= x or j <= y):
                img[i][j] = 255;
                
    return img
```

```python
# Calling the function #
img_translated = TranslateImage(100, 100, img)

# Save image #
cv.imwrite('skuna_translated.jpg', img_translated)

# Irá abrir uma aba com a imagem #
cv.imshow('sukuna translated', img_translated)

# necessário para não crashar o kernel do python #
cv.waitKey(0)
cv.destroyAllWindows()
```

## **5. Applying a Negative filter**

In this topic we'll be creating a negative filter and create a function to apply in a specified reagion of the image.

```python
# Getting image in gray scale #
img = cv.imread('images/sukuna.jpg', 0)

# Getting img shape #
shape = np.shape(img)
print("Image Height: {}\nImage Width: {}".format(shape[0], shape[1]))
```

```python
# Now we'll create a function to invert gray color

def negativeImg(img):
    
    shape = np.shape(img)
    
    for i in range(0, shape[0]):
        for j in range(0, shape[1]):
            img[i][j] = 255 - img[i][j]
        
    return img
```

```python
# applyig function #
img = negativeImg(img)

# Plotting the image #
cv.imshow('sukuna rotated', img)
cv.imwrite('images/sukuna_inverted.jpg', img)

# necessário para não crashar o kernel do python #
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# ploting image #
plt.subplots(figsize=(15, 8))

# Lendo a imagem #
img = plt.imread('images/sukuna_inverted.jpg', 0)
# Plotando a imagem #
plt.imshow(img, cmap = 'gray')
```
